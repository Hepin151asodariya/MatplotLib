###DATA VISUALIZATION TOOLS

BAR CHARTS(compares values between categaries)

```python
#plt.bar(x, height, color="color_name", width=value, label="label_name, alpha=value)
import matplotlib.pyplot as plt

product = ["A","B","C","D"]
sales=[1000,1500,800,1200]

plt.bar(product,sales,color="orange",label="sales 2025")
plt.xlabel("product")
plt.ylabel("sales")
plt.title("product sales comparison")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

product = ["A","B","C","D"]
sales=[1000,1500,800,1200]

#chart is convert into horizontally using plt.barh
plt.barh(product,sales,color="orange",label="sales 2025")
plt.xlabel("product")
plt.ylabel("sales")
plt.title("product sales comparison")
plt.legend()
plt.show()
```

PIE CHARTS(show a parts of whole)

```python
#plt.pie(values, labels=labels_list, colors=colres_list, autopct='%1.1f%%')
import matplotlib.pyplot as plt

regions = ["North","South","East","West"]
revenue = [3000,2000,1500,1000]

plt.pie(revenue,labels=regions,autopct="%1.1f%%",colors=["gold","skyblue","lightgreen","coral"])
plt.title("Revenue Contribution By Region")
plt.show()
```

HISTOGRAM(showing distribution of data)

```python
#plt.hist(data, bins=number_of_bins, color="color_name", edgecolor="edge_color")
import matplotlib.pyplot as plt

scores=[45,67,50,58,96,32,36,54,96,58,71,12,56,32,69]
plt.hist(scores,bins=5,color="purple",edgecolor="white",linewidth=2)
plt.xlabel("Score Range")
plt.ylabel("Number Of Students")
plt.title("Score Distribution Of Students")
plt.show()
```
