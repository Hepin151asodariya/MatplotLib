###SIMPLE PLOT CHART

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y =[10,30,20,40]

plt.plot(x,y)
plt.title("Simple Line Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

###MATPLOTLIB PLOTTING FUNCTION

```python
# plt.plot(x,y, color="colorname", linestyle="line_style", marker="marker_style", lable="Label_name")
import matplotlib.pyplot as plt

months = [1,2,3,4]
sales =[100,1000,200,400]

plt.plot(months,sales,color="blue",linestyle="--",marker="o",linewidth=2,label="2025 sales data")
plt.title("Monthly sales data report")
plt.xlabel("months")
plt.ylabel("sales")
plt.legend(loc="upper left")
plt.grid(color="gray",linestyle=":",linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],["M1","M2","M3","M4"])
plt.show()
```
